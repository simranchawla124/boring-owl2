# Popularity Based Recommender System 

```python
ratings.head()
#the dataset ratings has User Id, Book iSBN, and rating given by the user
#not all books have been given raating, there may be some books without rating
#not all users give ratings some may not
```

```python
books.head()
#the books dataset conatains book info such as book unique id(isbn), title,authorname, publishername and image urls of the book
```

```python
# We perform a join operation on books and ratings in order to get the avg rating of the books
# Here in popularity based recommendation system instead of just taking the criteria of highest avg rating we also take a min no of votes criteria
# as a book may have only two ratings but both 10 which user might not prefer so we take a min votes of 250 on the book
```

```python
#we store the book and it's ratings in books_plus_rating
books_plus_rating=books.merge(ratings,on='ISBN')
```

```python
# After merging we group the books based on the book title as different editions of a book may have different ISBN's
# and we calculate the count of votes given to each book
# a series is given as output but to display it in dataframe format we use to_frame method
# and then we reset the indices to 0,1,2 instead of book titles
```

```python
#we save the no of votes per book in no num_of_ratingspb
num_of_ratingspb=books_plus_rating.groupby('Book-Title').count()['Book-Rating'].to_frame().reset_index()
```

```python
#for better readability we rename the column to No-of-votes instead of book-rating and inplace=true makes the changes permenant
num_of_ratingspb.rename(columns={'Book-Rating':'No_of_votes'},inplace=True)
```

```python
num_of_ratingspb
```

```python
# in this we calculate average rating/book by using mean function
avg_ratingpb=books_plus_rating.groupby('Book-Title').mean()['Book-Rating'].to_frame().reset_index()
avg_ratingpb.rename(columns={'Book-Rating':'Avg Rating'},inplace=True)
avg_ratingpb
```

```python
#here we merge the dataframes avg rating and num of ratings
popularity_index=num_of_ratingspb.merge(avg_ratingpb,on='Book-Title')
```

```python
popularity_index
```

```python
# here we use loc to get all the rows whose no_of_votes are greater than 250
# we chose only those books who have minimum votes >250 because many books have high average rating but no of ratings are less
print(popularity_index['No_of_votes'].min())
print(popularity_index['No_of_votes'].max())
print(popularity_index['No_of_votes'].mean())
print(popularity_index['No_of_votes'].median())
print(popularity_index['No_of_votes'].mode())
popularity_index=popularity_index.loc[popularity_index.No_of_votes>250]
```

```python
popularity_index
```

```python
# here when we get the books with >250 votes we sort them accordting toavg rating to get top rating books
popularity_index=popularity_index.loc[popularity_index.No_of_votes>250].sort_values('Avg Rating',ascending=False)
```

```python
popularity_index
```

```python
# we get the top 50 books by using head()
top_rated_50=popularity_index.head(50)
```

```python
top_rated_50
```

```python
#checking the contents of books dataset
books.head()
```

```python
books.shape
```

```python
#we merge the top_rated_50 with books to get url anf publisher name and all for final publishing on UI
# but we need to drop duplicates as with the same title but different isbn's book may be present in books dataset
top_rated_50.merge(books,on='Book-Title')
```

```python
# due to multiple isbn numbers for one book we get duplicates
top_rated_50.merge(books,on='Book-Title').drop_duplicates('Book-Title')
```

```python
top_rated_50
```

```python
# upgraded version of recommend function
def recommend2(book_name):
    # we need to fetch the book index from book_name
    index=np.where(pt.index.str.lower()== book_name.lower())[0][0]
    distances=similarity[index]
    similar_items=sorted(list(enumerate(distances)),key=lambda x:x[1],reverse=True)[1:6]
    #print(similar_items)
    data=[]
    for i in similar_items:
        #print(pt.index[i[0]])
        items=[]
        temp=books[books['Book-Title']==pt.index[i[0]]]
        items.extend(temp.drop_duplicates('Book-Title')['Book-Title'])
        items.extend(temp.drop_duplicates('Book-Title')['Book-Author'])
        items.extend(temp.drop_duplicates('Book-Title')['Image-URL-M'])
        data.append(items)
    return data
        #return the title which is similar to the index number that we got from similar items
    #return suggestions
```
